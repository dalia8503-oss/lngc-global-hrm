```python
import pandas as pd
import numpy as np

# 핵심 기능 요약
df = pd.read_csv('data/current_normal.csv')  # 데이터 불러오기
df.head()         # 상위 5행 확인
df.info()         # 데이터 정보
df.describe()     # 기술통계
df.shape          # (행, 열)
df.dtypes         # 자료형 확인
df['Time']      # 특정 컬럼 선택
df[['0','1.91','3.81','5.72']]  # 여러 컬럼 선택
```

```python
# 행 추가 (concat)
df1 = pd.DataFrame({'이름': ['A','B'], '나이': [20,30]})
df2 = pd.DataFrame({'이름': ['C','D'], '나이': [25,35]})
df_concat = pd.concat([df1, df2], ignore_index=True)
print(df_concat)

# 열 기준 병합 (merge) - SQL의 JOIN과 유사
df_info = pd.DataFrame({'이름': ['A','B','C'], '부서': ['개발','기획','디자인']})
df_salary = pd.DataFrame({'이름': ['A','B','D'], '급여': [5000,4500,4800]})

# inner join (공통된 것만)
inner = pd.merge(df_info, df_salary, on='이름', how='inner')
# left join (왼쪽 기준)
left = pd.merge(df_info, df_salary, on='이름', how='left')
# outer join (전체)
outer = pd.merge(df_info, df_salary, on='이름', how='outer')

print("inner")
print(inner)
print("left")
print(left)
print("outer")
print(outer)
```

```python
# 단일 기준 정렬
left.sort_values('이름')                # 오름차순
left.sort_values('이름', ascending=False)  # 내림차순

# 다중 기준 정렬
left.sort_values(['이름', '부서'], ascending=[True, False])

# 인덱스 기준 정렬
left.sort_index()
```

```python
df = pd.merge(left, df1, on='이름', how='left')

# 단일 조건
df[df['나이'] >= 30]            # 나이 30 이상
df[df['부서'] == '개발']        # 부서가 '개발'인 행

# 복합 조건 (& : AND, | : OR)
df[(df['나이'] >= 25) & (df['급여'] >= 4000)]
df[(df['부서'] == '개발') | (df['부서'] == '기획')]

# isin(): 여러 값 중 하나와 일치
df[df['부서'].isin(['개발', '기획'])]

# loc[행 조건, 열 선택]
df.loc[df['나이'] >= 30, ['이름', '급여']]

# iloc[행 번호, 열 번호]
df.iloc[0:3, 1:3]   # 0~2행, 1~2열
```
